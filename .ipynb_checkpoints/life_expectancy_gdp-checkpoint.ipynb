```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data.csv')  # Load data
print(df.head())  # Examine what the data looks like
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('Change in GDP over Time per Country')
plt.savefig('GDP_time.png')
plt.show()
plt.clf()
```

# From the figure above, we see the US consistantly having the highest GDP. We also see China has the largest increase in GDP from 2000 to 2014. Moreover, the US and China are the only two countries with a singificant increase in GDP in those years. The GDP of Germany, Mexico, Chile, and Zimbabwe  did not change very much

```python
plt.figure(figsize=(12,7))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Change in Life Expectancy over Time per Country')
plt.savefig('LE_time.png')
plt.show()
plt.clf()
```

# From the fiure above, we see that for Chile, China, Germany, Mexico, and the US, life expectancy increased at roughly equal rates between 2000 and 2014. However, the story is different for Zimbabwe, which started much lower than the rest, initially decreased a little, and ultimately increased the most, but still ended as the lowest.

```python
plt.figure(figsize=(12,7))
sns.barplot(data=df, x='Country', y='Life expectancy at birth (years)', estimator=np.median)
plt.title('Median Life Expectancy for Each Country (2000-2015)')
plt.savefig('LF_Cntry.png')
plt.show()
plt.clf()
```

# The figure above shows that Chile, China, Germany, Mexico, and the US have had roughly the median same life expectancy from 2000 to 2014 (around 74-80), while Zimbabwe had the lowest (45-54).

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=df[df['Life expectancy at birth (years)'] > 70], x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy vs GDP for Different Countries')
plt.savefig('LE_GDP.png')
plt.show()
plt.clf()
```

# Here, we see a positive (mostly linear) relationship between life expectancy and GDP. It is interesting to note that countries with a smaller GDP (Chile, Mexico, Germany, China at first) seem to increase the life expectancy at a higher rate when GDP increases.
